# Benchmark QRT

This notebook illustrates a simple benchmark example that should help novice participants to start the competition.

## Used libraries

```python
import seaborn as sns
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import KFold
import matplotlib.pyplot as plt
import statsmodels.api as sm
import random
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import StratifiedKFold
```

## Loading data

The train and test inputs are composed of 46 features.

The target of this challenge is `RET` and corresponds to the fact that the **return is in the top 50% of highest stock returns**.

Since the median is very close to 0, this information should not change much with the idea to predict the sign of the return.

```python
x_train = pd.read_csv('../x_train.csv', index_col='ID')
y_train = pd.read_csv('../y_train.csv', index_col='ID')
train = pd.concat([x_train, y_train], axis=1)
test = pd.read_csv('../x_test.csv', index_col='ID')
train.head()
```

```python
ret=[f'RET_{k}' for k in range(20,0,-1)]
vol=[f'VOLUME_{k}' for k in range(20,0,-1)]
```

# Exploration des secteurs

Nombre de points par secteur :

# Cleaning : remplacement des NaN par les mÃ©dianes par secteur

```python
train_cl = train.copy()
for sect in range(nb_sec):
    for ret_t,vol_t in zip(ret,vol):
        moy_ret = train[ret_t][train['SECTOR']==sect].median()
        moy_vol = train[vol_t][train['SECTOR']==sect].median()
        train_cl[ret_t] = train_cl[ret_t].fillna(moy_ret)
        train_cl[vol_t] = train_cl[vol_t].fillna(moy_vol)
train_cl.head()
```

# Scaling

```python
scaler = StandardScaler()
```

```python
train_cl[ret+vol] = scaler.fit_transform(train_cl[ret+vol])
train_cl.head()
```

# SÃ©lection des features

```python
nb_feat = 20
```

```python
features_base = ret[-nb_feat:]+vol[-nb_feat:]
target = 'RET'
```

## New features

```python
# Feature engineering

nb_shifts = 5 #nombre de features supplÃ©mentaires

new_features = []

# Conditional aggregated features
shifts = range(1,nb_shifts+1)  # Choose some different shifts
statistics = ['median']  # the type of stat
gb_features = ['SECTOR', 'DATE']
target_feature = 'RET'
tmp_name = '_'.join(gb_features)
for shift in shifts:
    for stat in statistics:
        name = f'{target_feature}_{shift}_{tmp_name}_{stat}'
        feat = f'{target_feature}_{shift}'
        new_features.append(name)
        for data in [train_cl]:
            data[name] = data.groupby(gb_features)[feat].transform(stat)
```

```python
features = features_base + new_features
```

```python
features
```

# ModÃ¨les Ã  la zob

```python
N = 5 #nombre de folds dans la cross validation

for sect in range(nb_sec):
    x_train_cl = train_cl[features][train_cl['SECTOR']==sect]
    y_train_cl = train_cl[target][train_cl['SECTOR']==sect]
    
    rf_params = {
    'n_estimators': 500,
    'max_depth': 2**3,
    'random_state': 0,
    'n_jobs': -1
    }
    
    model = RandomForestClassifier(**rf_params)
    skf = StratifiedKFold(n_splits=N,shuffle=True)
    results = cross_val_score(model,x_train_cl,y_train_cl,cv=skf,scoring='accuracy')
    
    for fold in range(N):
        print(f"Secteur {sect} - Fold {fold+1} - Accuracy: {results[fold]*100:.2f}%")
    print(f"Secteur {sect} - Accuracy: {np.mean(results)*100:.2f}%")
```

# ModÃ¨les corrigÃ©s

```python
N = 4 #nombre de folds

#train_cl = train_cl.sample(n=int(3e4),random_state=0) # DÃ©commenter pour travailler sur un Ã©chantillon

splits = StratifiedKFold(n_splits=N, random_state=1,shuffle=True).split(train_cl[features],train_cl[target])

rf_params = {
'n_estimators': 500,
'max_depth': 2**3,
'random_state': 0,
'n_jobs': -1
}

scores = []

for i, (ind_train,ind_test) in enumerate(splits):
    models = []
    x_test_cl = pd.DataFrame(columns=features)
    y_test_cl = pd.DataFrame(columns=[target])
    y_pred = pd.DataFrame(columns=[target])
    index_train = train_cl.iloc[ind_train].index
    index_test = train_cl.iloc[ind_test].index
    
    for sect in range(nb_sec):
                
        x_train_cl = train_cl[train_cl['SECTOR']==sect].loc[index_train,features].dropna()
        y_train_cl = train_cl[train_cl['SECTOR']==sect].loc[index_train,target].dropna().astype('bool')
        
        model = RandomForestClassifier(**rf_params)
        model.fit(x_train_cl, y_train_cl)
        models.append(model)
        
        print(f'ModÃ¨le {sect} entraÃ®nÃ©')
        
        new_x_test = train_cl[train_cl['SECTOR']==sect].loc[index_test,features].dropna()
        x_test_cl = pd.concat([x_test_cl,new_x_test])
        new_y_test = train_cl[train_cl['SECTOR']==sect].loc[index_test,target].dropna().astype('bool')
        y_test_cl = pd.concat([y_test_cl,new_y_test])
        
        y_pred = pd.concat([y_pred,pd.Series(model.predict_proba(new_x_test)[:,-1],index=new_x_test.index)])

    sub = train_cl.loc[index_test].copy()
    sub['pred'] = y_pred.iloc[:,0]
    y_pred = sub['pred'].transform(lambda x: x > x.median()).values
    
    score = accuracy_score(y_test_cl.iloc[:,0].astype('bool'), y_pred)
    scores.append(score)
    print(f"Fold {i+1} - Accuracy: {score* 100:.2f}%")

mean = np.mean(scores)*100
std = np.std(scores)*100
u = (mean + std)
l = (mean - std)
print(f'Accuracy: {mean:.2f}% [{l:.2f} ; {u:.2f}] (+- {std:.2f})')
```
